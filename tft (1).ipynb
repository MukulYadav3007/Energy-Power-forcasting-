```python
import numpy as np

def sequence_generator(df, target_column, feature_columns, window_size, batch_size):
    num_samples = len(df) - window_size

    while True:
        for i in range(0, num_samples, batch_size):
            # Extract batch of data
            batch_df = df.iloc[i:i+batch_size+window_size]

            sequences = []
            targets = []

            for j in range(len(batch_df) - window_size):
                # Extract a sequence of length window_size
                seq = batch_df[feature_columns].iloc[j:j+window_size].values
                # Extract the target variable (next observation)
                target = batch_df[target_column].iloc[j+window_size]
                sequences.append(seq)
                targets.append(target)

            yield np.array(sequences), np.array(targets)


feature_columns = ['Electricity_consumption(kWh)', 'Avg_Voltage(Volt)', 'Avg_Current(Amp)', 'Freq(Hz)']

target_column = 'Electricity_consumption(kWh)'
# Define window size and batch size
window_size = 10
batch_size = 32  # Adjust as needed

# Create sequence generator
sequence_gen = sequence_generator(df, target_column, feature_columns, window_size, batch_size)

# Example usage:
# Iterate over the generator to get batches of sequences and targets
for i in range(3):  # Iterate for 3 batches as an example
    sequences, targets = next(sequence_gen)
    print("Batch", i+1)
    print("Shape of sequences:", sequences.shape)
    print("Shape of targets:", targets.shape)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate, MultiHeadAttention
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Reshape

class TemporalFusionTransformer(tf.keras.Model):
    def __init__(
        self,
        num_encoder_steps,
        num_decoder_steps,
        input_size,
        output_size,
        num_heads,
        embed_size=32,
        dense_layer_size=64,
        dropout=0.1,
        **kwargs
    ):
        super(TemporalFusionTransformer, self).__init__(**kwargs)
        self.num_heads = num_heads
        self.input_size = input_size
        self.embed_size = embed_size
        self.num_decoder_steps = num_decoder_steps
        self.num_encoder_steps = num_encoder_steps
        self.output_size = output_size
        self.dense_layer_size = dense_layer_size

        # Static and temporal embeddings
        self.static_embedding = Dense(embed_size)
        self.temporal_embedding = Dense(embed_size)
        self.attention_layers = self.create_attention_layers()
        self.encoder = self.create_encoder()
        self.decoder = self.create_decoder()

    def create_attention_layers(self):
        attention_layers = []
        for _ in range(self.num_decoder_steps):
            layers_list = []
            for _ in range(self.num_heads):
                layers_list.append(
                    MultiHeadAttention(
                        num_heads=self.num_heads,
                        key_dim=self.embed_size,
                        dropout=0.1,
                    )
                )
            attention_layers.append(layers_list)
        return attention_layers

    def create_encoder(self):
        input_feats = Input(shape=(self.num_encoder_steps, self.input_size))

        static_feats = self.static_embedding(input_feats)
        temporal_feats = self.temporal_embedding(input_feats)

        encoder_input = Concatenate()([static_feats, temporal_feats])
        encoder_output = Dense(self.dense_layer_size, activation="relu")(encoder_input)

        return Model(input_feats, encoder_output)



    def create_decoder(self):
        decoder_input = Input(shape=(self.num_encoder_steps, self.dense_layer_size))
        output = decoder_input

        # Reshape the input to match the expected shape
        output = Reshape((self.num_encoder_steps, self.dense_layer_size))(output)

        for attention_layers in self.attention_layers:
            for layer in attention_layers:
                output = layer(output, output)

        decoder_output = Dense(self.output_size)(output)
        return Model(decoder_input, decoder_output)



    def call(self, inputs):
        encoder_input, decoder_input = inputs

        encoder_output = self.encoder(encoder_input)

        # Reshape the decoder input to match the expected shape
        decoder_input = tf.reshape(decoder_input, [-1, self.num_encoder_steps, self.dense_layer_size])

        decoder_output = self.decoder(decoder_input)

        return decoder_output


# Define hyperparameters
NUM_ENCODER_STEPS = 10
NUM_DECODER_STEPS = 1
INPUT_SIZE = 4  # Adjust based on the number of features in your data
OUTPUT_SIZE = 1
NUM_HEADS = 2

# Create model instance
model = TemporalFusionTransformer(
    num_encoder_steps=NUM_ENCODER_STEPS,
    num_decoder_steps=NUM_DECODER_STEPS,
    input_size=INPUT_SIZE,  # Adjusted input size
    output_size=OUTPUT_SIZE,
    num_heads=NUM_HEADS
)

# Compile the model
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')

# Train the model
batch_size = 16
epochs = 50
model.fit([sequences, sequences[:, :NUM_ENCODER_STEPS, :]], targets, batch_size=batch_size, epochs=epochs)

# Make predictions
predictions = model.predict([test_sequences, test_sequences[:, :NUM_ENCODER_STEPS, :]])
```
